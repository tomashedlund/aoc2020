```python
i = 0
count = 0
for row in data.content:
    if row[i] == '#':
        count += 1
        row = row[:i] + 'X' + row[i+1:]
    else:
        row = row[:i] + 'O' + row[i+1:]
    i += 3
    if i > len(row)-1:
        i -= len(row)
    print(row, len(row), count)
print(count)
```

```python
i = [0, 0, 0, 0, 0]
count = [0, 0, 0, 0, 0]
step = [1, 3, 5, 7, 1]
for row in range(len(data.content)):
    for j in range(len(i)):
        if data.content[row][i[j]] == '#' and (j != 4 or row%2 == 0):
            count[j] += 1
        if j != 4 or row%2 == 0:
            i[j] += step[j]
        if i[j] > len(data.content[row])-1:
            i[j] -= len(data.content[row])
print(count)
countSum = 1
for c in count:
    countSum *= c
print(countSum)
```
