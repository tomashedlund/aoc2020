```python
count = 0
for s in data:
    l = set(''.join(s).replace (';', ''))
    print(l, len(l))
    count += len(l)
print(count)
```

```python
count = 0
for s in data:
    l = ''.join(s).replace (';', '')
    print(s[0],s)
    length = len(s)
    for x in s:
        if len(x) == 0:
            length -= 1
    for c in s[0]:
        if l.count(c) == length:
            count += 1
            print(c, l.count(c), length, "+1")
        else:
            print(c, l.count(c), length)
    print('\n')
print(count)
```
