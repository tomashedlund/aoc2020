```python
import pandas as pd
import re
import numpy as np
```

```python
f = open("C:\\Users\\TOHED\\Desktop\\day4.txt", "r")
data = f.read()
f.close()
```

```python
data = data.replace('\n', ';').split(";;")
```

```python
data = [re.split(";| ", line) for line in data]
#data = pd.DataFrame(data)
```

```python
data
```

```python
data = pd.DataFrame([dict(i.split(":") for i in list(filter(None, l))) for l in data])
```

```python
data
```

```python
data = data[data[['byr', 'iyr', 'eyr', 'hgt', 'hcl', 'ecl', 'pid']].notnull().all(1)]
data = data.loc[(data['byr'].astype('int32') >= 1920) & (data['byr'].astype('int32') <= 2002)]
data = data.loc[(data['iyr'].astype('int32') >= 2010) & (data['iyr'].astype('int32') <= 2020)]
data = data.loc[(data['eyr'].astype('int32') >= 2020) & (data['eyr'].astype('int32') <= 2030)]
```

```python
data.hgt = [x.replace('c', ' c') for x in data.hgt]
data.hgt = [x.replace('i', ' i') for x in data.hgt]
```

```python
data[['hgt', 'unit']] = data['hgt'].str.split(' ',expand=True)
```

```python
data = data[data[['unit']].notnull().all(1)]
```

```python
data2 = data[(data['unit'] == 'cm') & ((data['hgt'].astype('int32') < 150) | (data['hgt'].astype('int32') > 193))].index
data.drop(data2, inplace=True)
```

```python
data2 = data[(data['unit'] == 'in') & ((data['hgt'].astype('int32') < 59) | (data['hgt'].astype('int32') > 76))].index
data.drop(data2, inplace=True)
```

```python
data2 = data[(data['hcl'].str.len() != 7)].index
data.drop(data2, inplace=True)
```

```python
chars = set('0123456789abcdef')
data2 = []
for index, row in data.iterrows():
    if row.hcl[0] != '#':
        data2.add(row)
        print("#",row.hcl)
        continue
    
    s = row.hcl.replace("#", "", 1)
    if any((c not in chars) for c in s):
        data2.add(row)
        print(c,row.hcl)
        continue
data.drop(data2, inplace=True)
```
