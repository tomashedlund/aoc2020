```python
data = data.content.str.split('-|: | ',expand=True).add_prefix('column_')
```

```python
data.head()
```

```python
count = 0
for index, row in data.iterrows():
    c = row.column_3.count(row.column_2) 
    if c >= int(row.column_0) and c <= int(row.column_1):
        count += 1
        print(row, c)
print(count)
```

```python
count = 0
for index, row in data.iterrows():
    if (row.column_2 == row.column_3[int(row.column_0)-1]) != (row.column_2 == row.column_3[int(row.column_1)-1]):
        count += 1
        print(row.column_2, row.column_0, row.column_3[int(row.column_0)-1], row.column_1, row.column_3[int(row.column_1)-1], row.column_3)
print(count)
```
